```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

# Request page from wiki using requests

```python
headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64; rv:50.0) Gecko/20100101 Firefox/50.0'}
url = "https://en.wikipedia.org/wiki/List_of_Grammy_Award_winners_and_nominees_by_country"
page = requests.get(url, headers=headers)
```

```python
# print(page.text)
```

```python
pa = BeautifulSoup(page.text, "lxml")
```

```python
# print(pa.prettify())
```

# Title of this page

```python
### This is what you see on the tab ###
print(pa.title.getText())
```

# Extract all hyperlinks from the page

```python
all_as = pa.find_all('a', href=True)
```

```python
for a_tag in all_as:
    
    # all the text on the hyperlinks
    print(a_tag.get_text())
    
    # actual hyperlinks
    print(urljoin(url, a_tag['href']))
```

## Extract certain hyperlinks from the page

```python
# Find the parent node which contains all countries
u1 = pa.find('ul')

# Follow the previous steps
all_as_in_u1 = u1.find_all('a', href=True)

for a_tag in all_as_in_u1:
    
    # all the text on the hyperlinks
    print(a_tag.get_text())
    
    # actual hyperlinks
    print(urljoin(url, a_tag['href']))
```

# Scrap a table and store it as a dataframe

```python
# Retrieve the table element
table1 = pa.find_all('table')
# Second way
table2 = pa.find_all(class_="wikitable")
# Third way
table3 = pa.select("div table")

my_tab = table1[4]
```

## Table Headings

```python
headings = my_tab.find_all("th")

headings_ls = []

for h in headings:
    headings_ls.append(h.get_text())

headings_ls
```

## Table Content

```python
trs = my_tab.find_all("tr")

# Delete the headings from the list
trs.pop(0)

ls = []
for tr in trs:
    tds = tr.find_all('td')
    row = [td.text for td in tds]
    ls.append(row)
```

```python
pd.DataFrame(ls, columns=headings_ls)
```
